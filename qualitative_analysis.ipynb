# Fireflies synchronously flash

$$
\frac{\partial \phi}{\partial t} = \frac{2\pi}{T T_{0}}(T_{0} - T) - \beta \sin(\phi)
$$

$\phi$ is the phase difference between the flash periodic rhythm of a firefly and the stimulus of another firefly,
$T_{0}$ is the period of flashing of the firefly and $T$ is of the stimulus, and $\beta$ measures the relative influence of the zeitbeger.

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Equilibrium - Fixed Point Analysis


Fixed points represent equilibrium solutions. An equilibrium is a state where at $t$ approaches $\inf$, the system converges to the equilibrium. In order to find an equilibrium of equation (3) we set $\dot{\phi}$ to zero. 
    
$$
\frac{\partial \phi}{\partial t} = \frac{2\pi}{T T_{0}}(T_{0} - T) - \beta \sin(\phi) \stackrel{set}{=} 0\\
\frac{2\pi}{T T_{0}\beta}(T_{0} - T) = \sin(\phi_{e})\\
$$

Let $c = \frac{2\pi}{T T_{0}\beta}(T_{0} - T)$ which is a constant where $c \neq 0$. Then, we have 3 cases:

* case #1:

If $-1 < c < 1$ and $c \neq 0$: there are two fixed points. 

* case #2:

If $c > 1$ or $c < -1$: there are no fixed points. 

* case #3:

If $c = 1$ or $c = -1$: there is one fixed point. 

```python
x_long = np.linspace(-np.pi, np.pi, 1000)
x_short = np.linspace(-np.pi, np.pi, 10)
```

```python
plt.figure(figsize=(15, 10))

_ = plt.plot(x_long, np.sin(x_long), label="$f(\phi) = \sin(\phi)$")

_ = plt.plot(x_short, np.ones(len(x_short))*0.5, "--", c="r", label="$g(\phi) = 0.5$")
_ = plt.plot(x_short, np.ones(len(x_short))*-0.5, "--", c="r", label="$g(\phi) = -0.5$")

_ = plt.plot(x_short, np.ones(len(x_short))*1, "--", c="g", label="$g(\phi) = 1$")
_ = plt.plot(x_short, np.ones(len(x_short))*-1, "--", c="g", label="$g(\phi) = -1$")

_ = plt.plot(x_short, np.ones(len(x_short))*1.5, "--", c="y", label="$g(\phi) = 1.5$")
_ = plt.plot(x_short, np.ones(len(x_short))*-1.5, "--", c="y", label="$g(\phi) = -1.5$")

_ = plt.scatter(np.pi/2, 1, c="k")
_ = plt.scatter(-np.pi/2, -1, c="k")

_ = plt.xlabel("$\phi$", fontsize=20)
_ = plt.legend()
```

```python
# Python program to implement Runge Kutta method 
# A sample differential equation "dy / dx = (x - y)/2" 
def dydx(x, y): 
    T = 2*np.pi/1.01
    T0 = 2*np.pi
    b = 0.17
    return 2*np.pi*(T-T0)/(T*T0) - b*np.sin(y)



# def dydx(x, y): 
#     T = 2*np.pi/1.2
#     T0 = 2*np.pi
#     omega = 2*np.pi/T
#     b = 0.17
#     return 2*np.pi/T0 + b*np.sin(omega * x - y)

  
# Finds value of y for a given x using step size h 
# and initial value y0 at x0. 
def rungeKutta(x0, y0, x, h): 
    # Count number of iterations using step size or 
    # step height h 
    n = (int)((x - x0)/h)  
    # Iterate for number of iterations 
    y = y0 
    
    # return vector
    vec = np.zeros(n+1)
    vec[0] = y0
    
    for i in range(1, n + 1): 
        """Apply Runge Kutta Formulas to find next value of y"""
        k1 = h * dydx(x0, y) 
        k2 = h * dydx(x0 + 0.5 * h, y + 0.5 * k1) 
        k3 = h * dydx(x0 + 0.5 * h, y + 0.5 * k2) 
        k4 = h * dydx(x0 + h, y + k3) 
  
        # Update next value of y 
        y = y + (1.0 / 6.0)*(k1 + 2 * k2 + 2 * k3 + k4) 
  
        # Update next value of x 
        x0 = x0 + h 
        
        # update vector
        vec[i] = y
    return vec 
  
# Driver method 
t0 = 0
y = 1.2
tf = 700
h = 0.2
vec1 = np.mod(rungeKutta(t0, y, tf, h), (T0)*np.ones(int(tf/h + 1)))
vec2 = np.mod(rungeKutta(t0, 0.5, tf, h), np.pi*2*np.ones(int(tf/h + 1)))
vec3 = np.mod(rungeKutta(t0, -0.5, tf, h), np.pi*2*np.ones(int(tf/h + 1)))
```
